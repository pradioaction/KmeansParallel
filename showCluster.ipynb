```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python

df1 = pd.read_csv("./output_centro_prallel.csv")
df1 = df1.loc[df1.x!=-1]
```

```python

plt.figure(dpi=150,figsize=(8,8))

# my_x_ticks = np.arange(5000, 11000, 1000)
#对比范围和名称的区别
#my_x_ticks = np.arange(-5, 2, 0.5)
my_y_ticks = np.arange(5000, 11000, 1000)
# plt.xticks(my_x_ticks)
plt.yticks(my_y_ticks)
sns.scatterplot(x=df1.x, y=df1.y ,
                # hue=df1.index, 
                palette=sns.color_palette("hls", n_colors=5),markers=".",size=1)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Clustered")

plt.show()
```
